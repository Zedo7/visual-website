```python
import numpy as np
import pandas as pd
df = pd.read_csv("Downloads/games.csv").drop(columns = ["Unnamed: 0"])
```

```python
df1 = df[["Title","Release Date","Rating"]].copy()
df1['Release Date'] = df1['Release Date'].replace('releases on TBD', 'Jan 01, 9999')
df1['Year'] = df1['Release Date'].str.split(" ").str.get(2).astype(int)
#df1.to_csv('~/dsc 106/q1data.csv', index=False)
```

```python
df1
```

```python
# Convert 'Year' column to integer
df1['Year'] = df1['Year'].astype(int)
df1 = df1[(df1['Year'] >= 1980) & (df1['Year'] <= 2023)]

# Define rating buckets
bins = [0, 1, 2, 3, 4, 5]
labels = ['0-1', '1-2', '2-3', '3-4', '4-5']

# Assign each game to a rating bucket
df1['RatingBucket'] = pd.cut(df1['Rating'], bins=bins, labels=labels, include_lowest=True)

# Count the number of games in each rating bucket per year
df_rating_counts = df1.groupby(['Year', 'RatingBucket']).size().unstack(fill_value=0)

# Reset the index to make 'Year' a column again, if necessary
df_rating_counts.reset_index(inplace=True)
df_rating_counts
df_rating_counts.to_csv('~/dsc 106/game_ratings_per_year.csv', index=False)
```

```python
from fuzzywuzzy import fuzz, process

def find_best_match(game):
    # Find best match in df1['Title'] for the given game
    match = process.extractOne(game, df1['Title'], scorer=fuzz.token_sort_ratio)
    # Return the match if the score is above 90
    return match[0] if match[1] > 90 else None

# Apply the function to the 'Title' column of df_awards
df_awards['Matched Title'] = df_awards['Title'].apply(find_best_match)

# Merge df1 and df_awards on the matched titles
df_awards = pd.merge(df_awards, df1[['Title', 'Rating']], left_on='Matched Title', right_on='Title', how='left')

# Drop unnecessary columns
df_awards = df_awards.drop(['Matched Title', 'Title_y'], axis=1)

# Rename columns for clarity
df_awards = df_awards.rename(columns={'Title_x': 'Title', 'Rating': 'Awarded Game Rating'})

df_awards
```

```python
import pandas as pd
import ast
from collections import Counter
from typing import List

# Convert 'Release Date' to datetime
df['Release Date'] = df['Release Date'].replace('releases on TBD', 'Jan 1, 0000')
df['Release Date'] = pd.to_datetime(df['Release Date'])
df['Year'] = df['Release Date'].dt.year

def count_genres_by_year(df):
    # Dictionary to hold genre count for each year
    yearly_genre_counter = {}

    for index, row in df.iterrows():
        year = row['Year']
        genres = row['Genres']

        # Initialize a Counter for this year if it does not exist
        if year not in yearly_genre_counter:
            yearly_genre_counter[year] = Counter()

        # Update the counter with genres
        yearly_genre_counter[year].update(genres)

    return yearly_genre_counter

yearly_genre_counter = count_genres_by_year(df)

# Convert dictionary of Counters to a DataFrame
genre_df = pd.DataFrame.from_dict(yearly_genre_counter, orient='index')

# Transpose so that genres are rows and years are columns
genre_df = genre_df.transpose()


# Write to csv
genre_df = genre_df.fillna(0)
genre_df = genre_df.drop(columns=[2024, 2025])
genre_df.to_csv('~/dsc 106/q2data.csv')
```

```python
df3['Year'] = df3['Year'].fillna(df3['Year'].mean()).astype(int)

# Now, we'll group the data by Year and sum up the sales in each region
sales_per_year = df3.groupby('Year').agg({
    'NA_Sales': 'sum',
    'EU_Sales': 'sum',
    'JP_Sales': 'sum',
    'Other_Sales': 'sum',
}).reset_index()

# Let's view the new dataframe
print(sales_per_year)
```
